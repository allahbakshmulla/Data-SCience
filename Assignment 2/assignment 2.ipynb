```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('imdb_top_250_movies.csv')
```

```python
data
```

```python
sorted=data.sort_values(by=['rating'],ascending=False)
```

```python
sorted
```

```python
rating = list()
```

```python
rating = sorted['rating'].tolist()
```

```python
rating
```

```python
names = sorted['movie_title'].tolist()
```

```python
names
```

```python
topMovies=list()
theirRating=list()
```

```python
for i in range(10):
    topMovies.append(names[i])
    theirRating.append(rating[i])
```

```python
topMovies,theirRating
```

```python

plt.xlabel('Movie name')
plt.ylabel('Rating')
plt.barh(topMovies,theirRating)
plt.savefig('result.png')
```
